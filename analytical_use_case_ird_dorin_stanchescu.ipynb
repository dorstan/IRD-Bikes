```python
df['Discount'].replace(np.nan, 0, inplace=True)
df['Discount USD'].replace(np.nan, 0, inplace=True)

df.isnull().sum()
```

```python
df1 = df[['OrderNumber', 'ProdDescr', 'SalesQuantity']]
df1.head()
```

```python
from itertools import permutations,combinations
from collections import Counter
```

```python
# Function that creates a list with pairs of items

def pairs (value):
    pairs = pd.DataFrame(list(permutations(value.values,2)),columns=["Prod1","Prod2"])
    return pairs
```

```python
# Group the products by OrderNumber then apply the function 

df_pairs = df1.groupby('OrderNumber')['ProdDescr'].apply(pairs).reset_index(drop=True)
df_pairs.head()
```

```python
# Check intervals of pairs

df_pairs_int = df_pairs.groupby(['Prod1', 'Prod2']).size()
df_pairs_int
```

```python
# Create a sorted dataframe by Desc

sorted_df = df_pairs_int.reset_index()
sorted_df.columns = ['Prod1', 'Prod2', 'Frequency']
sorted_df.sort_values(by='Frequency', ascending = False, inplace = True)
sorted_df.head()
```

```python
# Clean the dataset rows
cleaned_df = sorted_df[sorted_df["Prod1"]!=sorted_df["Prod2"]]
print(f'Number of rows: {cleaned_df.shape}')
cleaned_df.head()
```

```python
#Create a combination of groups so that can be used as an alternative to pairing
df2= df[['OrderNumber', 'ProdDescr']]
df2 = df2[df2['OrderNumber'].duplicated(keep=False)]
df2['Pairs'] = df2.groupby('OrderNumber')['ProdDescr'].transform(lambda x:','.join(x))
df2 = df2[['OrderNumber', 'Pairs']].drop_duplicates()
df2.head()
```

```python
# Count each row combination by the the pairing of the two


count = Counter()
for row in df2['Pairs']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

for key, value in count.most_common(10):
    if key[0] != key[1]:
        print(key, value)
```

```python
# Identify all pairings with Air Pump
air_pump_values = []
for key, value in count.most_common(20):
    result = [key[0], key[1], value]
    air_pump_values.append(result)
        
df_air = pd.DataFrame(air_pump_values, columns = [['Product1', 'Product2', 'Value']])
df_air
```
